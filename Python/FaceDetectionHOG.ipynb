```
# import libraries
import cv2
import numpy as np
import dlib
from imutils import face_utils
import matplotlib.pyplot as plt
from google.colab import files
```

```
def detectFaces(imgPath):
  
  # open image and store its RGB and GRAY
  image = cv2.imread(imgPath)
  im = np.float32(image)
  rgbImg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
  # calculate gradients and preprocessing
  gx = cv2.Sobel(gray, cv2.CV_32F, 1, 0, ksize=1)
  gy = cv2.Sobel(gray, cv2.CV_32F, 0, 1, ksize=1)
  mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
  
  # use detector
  faces = dlib.get_frontal_face_detector()
  
  # draw rectangles
  rects = faces(gray, 1)
  for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(rgbImg, (x, y), (x + w, y + h), (0, 255, 255), 10)
  
  # show results
  print('Number of detected faces:', len(rects))
  plt.imshow(rgbImg)
    
```
