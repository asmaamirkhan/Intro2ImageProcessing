```
# import libraries
import cv2
import numpy as np
import dlib
from imutils import face_utils
import matplotlib.pyplot as plt
from google.colab import files
```

```
def detectFaces(imgPath):
  
  # open image and store its RGB
  image = cv2.imread(imgPath)
  rgbImg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  
  # use detector
  cnn_face_detector = dlib.cnn_face_detection_model_v1('mmod_human_face_detector.dat')
  faces = cnn_face_detector(image, 1)
  
  # create rectangles
  rects = dlib.rectangles()
  rects.extend([d.rect for d in faces])

  # draw rectangles
  for (i, rect) in enumerate(rects):
    (x, y, w, h) = face_utils.rect_to_bb(rect)
    cv2.rectangle(rgbImg, (x, y), (x + w, y + h), (0, 255, 255), 10)
  
  # show results
  print('Number of detected faces:', len(rects))
  plt.imshow(rgbImg)
  
```
