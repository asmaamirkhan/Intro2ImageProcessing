```
# import libraries
import cv2
import matplotlib.pyplot as plt
from google.colab import files
```

```
def detectFaces(imgPath):
  # open classifier
  haar = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
  #haar = cv2.CascadeClassifier('haarcascade_frontalface_alt_tree.xml')
  #haar = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
  
  # open image and store its RGB and GRAY
  image = cv2.imread(imgPath)
  rgbImg = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  
  # find faces
  faces = haar.detectMultiScale(gray, scaleFactor=1.13, minNeighbors=4)
  
  # draw rectangles
  for i,(x,y,z,w) in enumerate(faces):
    cv2.rectangle(rgbImg, (x,y), (x+z,y+w), (0,256,256), 10)
    cv2.putText(rgbImg, str(i+1),(x, y), cv2.FONT_HERSHEY_SIMPLEX, 2,(255,0,0),5)
    
  # show image
  print('Number of detected faces:', len(faces))
  plt.imshow(rgbImg)
```
